```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#loading csv file as a data frame using pandas
df = pd.read_csv('cars.csv')
```

```python
df.dropna(inplace=True)
```

```python
df.columns
```

```python
passed_emissions=pd.get_dummies(df['passedemissions'],drop_first=True)
```

```python
passed_emissions
```

```python
passed_emissions.rename(columns={True:'Emission_test'}, 
                 inplace=True)
```

```python
passed_emissions
```

```python
df = pd.concat([df,passed_emissions],axis=1)
```

```python
df.head()
```

```python
df.drop(['passedemissions','carname'],axis=1,inplace=True)
```

```python
df.head()
```

```python
X=df.drop('Emission_test',axis=1)
y=df['Emission_test']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logm = LogisticRegression()
```

```python
logm.fit(X_train,y_train)
```

```python
estimations = logm.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,estimations))
```
